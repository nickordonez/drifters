For efficiency, this tutorial will rely on the Xarray to access and organize trajectory output from the ‘drifters’ (change name as needed) package for MOM6. Directions for running a Double Gyre test case can be found here (link).

In this example 100 evenly-spaced particles were initialized between 2 and 16 degrees latitude, and 32 and 46 degrees longitude. Total runtime for this example was 100 days. 

Load the necessary modules.

```python
import netCDF4 as nc
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
%matplotlib inline
```

The trajectory dataset is first loaded and assigned a variable. In some cases trajectories may be stored across multiple files, and this step may have to be repeated for each file. 

```python
ds = xr.open_dataset('drifter_trajectories.nc')
```

From here, trajectories for all particles can be plotted. 

```python
plt.scatter(ds.lon, ds.lat)
```

If one wishes to call specific particles, each particle must be assigned a global ID based on corresponding cell and count IDs.

```python
part_id = ds.id_cnt*(2**32)+ds.id_ij
```

One must then narrow the original dataset down to include information for a specific trajectory. In this case, we want to plot the trajectory for the particle in the very bottom-left corner of the plot of all trajectories.

```python
lon1 = ds.where(ds.lon <= 2.30, drop=True)
lon2 = lon1.where(lon1.lon >= 2.00, drop=True)
lat1 = lon2.where(lon2.lat <= 33.2, drop=True)
lat2 = lat1.where(lat1.lat >= 32.0, drop=True)
```

With the dataset now narrowed down, a global ID can be assigned to the particles within this set. Defining the type here will prevent rounding. 

```python
glob_id = lat2.id_cnt.astype(int)*(2**32)+lat2.id_ij.astype(int)
np.unique(glob_id)
```

With a list of particle IDs now given, one or more particle IDs can be selected and have their trajectories plotted using corresponding latitudes and longitudes.

```python
ds_part = (part_id == 4294967657)
plt.scatter(ds.lon[ds_part], ds.lat[ds_part])
```

Depending on the output of a run, other particle characteristics can also be plotted simultaneously. In the case of this run, particle velocities are also written to the trajectory file. 

First, one must define velocity or ‘speed’. Since velocities are given as zonal and meridional components, ‘velocity’ will be the magnitude.

```python
velocity = np.sqrt(ds.uvel**2 + ds.vvel**2)
ds['velocity'] = velocity 
velocity 
```

We can then plot the same trajectory as before, but with velocities included. 

```python
plt.scatter(ds.lon[ds_part], ds.lat[ds_part], c=ds.velocity[ds_part])
plt.colorbar()
```

The same can be done for the entire set of particles.

```python
plt.scatter(ds.lon, ds.lat, c=ds.velocity)
plt.colorbar()
```
